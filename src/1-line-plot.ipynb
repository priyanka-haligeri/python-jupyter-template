### Exercise 1: Line Plot Customization

```python
#
import matplotlib.pyplot as plt
import numpy as np

# Generate x values from 0 to 10
x = np.arange(0, 11)  # 0 to 10 inclusive

# Calculate y values as the square of x
y = x ** 2

# Create the plot
plt.plot(x, y, marker='o', color='blue', linestyle='-')

# Customize the plot
plt.title('Square of Numbers from 0 to 10')
plt.xlabel('X Values')
plt.ylabel('Y = X^2')
plt.grid(True)

# Display the plot
plt.show()

```
